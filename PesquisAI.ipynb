# 👨‍🔬 PesquisAI — Agente de IA para Pesquisadores Científicos

O **PesquisAI** é um assistente de inteligência artificial executado no Google Colab, projetado para automatizar o fluxo de trabalho acadêmico — desde a revisão bibliográfica até a preparação de artigos científicos.

### 🛡️ Vantagens

| # | Benefício | Detalhe |
|---|-----------|---------|
| 1 | **100% Gratuito** | Roda na infraestrutura em nuvem do Google |
| 2 | **Sem Instalação** | Pronto para uso direto no navegador |
| 3 | **Foco Nacional** | Integrado com **IBGE** e **DataSUS** |
| 4 | **Base Tecnológica** | Construído sobre a arquitetura **OpenCode** |

---

### 🚀 Como Iniciar

1. No menu superior, clique em **Ambiente de execução → Executar tudo** (ou `Ctrl + F9`).
2. Aguarde ~2 minutos enquanto as dependências e *skills* são instaladas.
3. Clique que ira aparecer e o PesquisAI abrirá em uma nova aba.

````python
import os, sys, subprocess, shutil

# Repositório com todo o código
REPO_URL = "https://github.com/gustavobraga-byte/PesquisAI.git"
WORK_DIR = "/tmp/pesquisai"


# Mostrar loading
print("⏳ Carregando o PesquisAI...")

# Limpar diretório anterior se existir
if os.path.exists(WORK_DIR):
    shutil.rmtree(WORK_DIR)

# Clonar repositório
print("📥 Baixando arquivos...")
subprocess.run(
    ["git", "clone", "--depth", "1", REPO_URL, WORK_DIR],
    capture_output=True
)

# Mudar para o diretório e executar
os.chdir(WORK_DIR)
sys.path.insert(0, WORK_DIR)

# Importar e rodar tudo
from main import run
run()
````
