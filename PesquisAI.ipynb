# 👨‍🔬 PesquisAI — Agente de IA para Pesquisadores Científicos

O **PesquisAI** é um assistente de inteligência artificial executado no Google Colab, projetado para automatizar o fluxo de trabalho acadêmico — desde a revisão bibliográfica até a preparação de artigos científicos.

| Benefício | Detalhe |
|-----------|---------|
| **100% Gratuito** | Roda na infraestrutura em nuvem do Google |
| **Sem Instalação** | Pronto para uso direto no navegador |
| **Foco Nacional** | Integrado com **IBGE**, **DataSUS** e + fontes brasileiras |
| **Base Tecnológica** | Construído sobre a arquitetura **OpenCode** |

````python
import os, sys, subprocess, shutil
from IPython.display import display, HTML, clear_output

REPO_URL = "https://github.com/gustavobraga-byte/PesquisAI.git"
WORK_DIR = "/tmp/pesquisai"

display(HTML('<p style="color:#4fc3f7;font-family:monospace">⏳ <b>PesquisAI</b> — iniciando...</p>'))

if os.path.exists(WORK_DIR):
    subprocess.run(["git", "-C", WORK_DIR, "pull", "--ff-only", "--depth", "1"],
        capture_output=True)
else:
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, WORK_DIR],
        capture_output=True)

os.chdir(WORK_DIR)
sys.path.insert(0, WORK_DIR)
clear_output(wait=True)
from main import run
run()
````
